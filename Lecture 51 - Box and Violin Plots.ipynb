```python
data1 = randn(100)
data2 = randn(100)
```

```python
sns.boxplot([data1,data2])
#this is the command to show a box plot.
```

```python
sns.boxplot([data1,data2],whis=np.inf)
#the command whis=np.inf shows the outliers of the box plot. 
```

```python
sns.boxplot(data2, vert=True)
#vert=True will set the box plot horizontally. 
```

```python
sns.boxplot(data2, vert=False)
#vert=False will set the box plot vertically. 
```

```python
#Normal Distribution
data1 = stats.norm(0,5).rvs(100)

#Two gamma dist. Concatenated together
data2 = np.concatenate([stats.gamma(5).rvs(50)-1,
                       stats.gamma(5).rvs(50)])
#box plot both data1 and data2
sns.boxplot([data1,data2],whis=np.inf)
```

```python
sns.violinplot([data1,data2])
#violin plot will show how unevely your data is distributed. 
```

```python
sns.violinplot(data2,bw=.01)
#this is the result of plugging bandwidth to .01 (bw=.01) 
```

```python
sns.violinplot(data1,inner='stick')
#inner='stick' is basicly like combining the violin plot with a rug plot. 
```

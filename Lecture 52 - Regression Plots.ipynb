```python
#The normal imports
import numpy as np
from numpy.random import randn
import pandas as pd

#Import the stats library from numpy
from scipy import stats

#These are the plotting modules and libraries we'll use:
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns

#Command so that plots appear in the iPython Notebook
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
sns.lmplot('total_bill','tip',tips)
#lmpot is the linear regression line. Very nice Visual. 
```

```python
sns.lmplot('total_bill','tip',tips,
          scatter_kws={'marker':'o','color':'indianred'},
          line_kws={'linewidth':1,'color':'blue'})
#this dictioary is an example of how to modify the scatter plot and regression line. 
```

```python
sns.lmplot('total_bill','tip',tips,order=4,
          scatter_kws={'marker':'o','color':'indianred'},
          line_kws={'linewidth':1,'color':'blue'})
#notice the order=4 fit. This is if you dont want linear dsta. 
```

```python
sns.lmplot('total_bill','tip',tips,fit_reg=False)
#fit_reg=False removes the regression line and only shows the scatter plot. 
```

```python
tips.head()
```

```python
tips['tip_pect']=100*(tips['tip']/tips['total_bill'])
```

```python
tips.head()
```

```python
sns.lmplot('size','tip_pect',tips)
# x axis is 'size' and y axis is 'tip_pect'
#
```

```python
sns.lmplot('size','tip_pect',tips, x_jitter=.15)
#jitter allows for better visual to the data being displayed. 
```

```python
sns.lmplot('size','tip_pect',tips,x_estimator=np.mean)
#this makes the scatter plot connected
```

```python
sns.lmplot('total_bill','tip_pect',tips,hue='sex',markers=['x','o'])
#notice what the hue does and the different 'o' and 'x' markers. 
```

```python
sns.lmplot('total_bill','tip_pect',tips,hue='day')
#notice the hue without any kind of marker setting. automatic hue settings are nice. 
```
